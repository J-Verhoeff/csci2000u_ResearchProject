## Task #1 

- find the number of apps for each each catagory
- graph in a pyplot pie chart

```python
# create method to count occurances of each catagory
def countOccurances(array):
    items = {} # intiallizes empty dictionary
    for i in range(len(array)):
        if (array[i] in items) == False:
            items[array[i]] = 1 # if not in dictionary initiallize count to 0
        else:
            items[array[i]] += 1 # if item is in dictionary increase count by 1
    return items
```

```python
catagories = countOccurances(list(apps["Category"]))
labels = list(catagories.keys()) # catagories
sizes = list(catagories.values()) # number of times each catagory appears
pyplot.title("Google Play Store Apps: Catagory Breakdown") # graph title
pyplot.pie(sizes, labels=labels, rotatelabels=True, radius=4); # create the pie plot with large radius so labels clearly show
```

### result

Based on this graph it appears that the largest catagories on the Google Play Store are Family, Games, and Tools while the rest of the catagories have comparitvly small number of apps on the app store.

## Task #2

- See if there is a corellation between the price of the app and the number of downloads
- find number of free apps vs the number of paid apps

```python
price_to_downloads = pd.DataFrame(apps.loc[:, ['Installs', 'Price']]) # condenses data set to just downloads and price
price_to_downloads.head()
```

```python
# method to count paid vs paid
def cost(array):
    price = {'Free': 0,
            'Paid': 0} # initiallize dicionary to count the number of apps
    for i in range(len(array)):
        if array[i] == '0':
            price['Free'] += 1
        else:
            price["Paid"] += 1
            
    return price
```

```python
# create pie chart to compare free vs paid
price = cost(list(price_to_downloads['Price']))
pyplot.title('Cost of Apps')
pyplot.pie(list(price.values()), labels=list(price.keys()), radius=1); # graph number of free vs paid apps
```

```python
# large number of free apps vs small number of paid apps
# compare average installs
# first need a function to convert a string into a float
def convert(word):
    if type(word) != str:
        return word
    result = ''
    numbers = '.0123456789'
    for i in word:
        if i in numbers:
            result += i
    if result != '':
        return float(result)
    return 0.0
```

```python
# now find average downloads for free apps and compare to average downloads for paid apps
free_apps = []
paid_apps = []
for i in range(len(price_to_downloads)):
    if price_to_downloads.iloc[i,1] == '0':
        free_apps.append(convert(price_to_downloads.iloc[i, 0]))
    else:
        paid_apps.append(convert(price_to_downloads.iloc[i,0]))
```

```python
xs = [np.mean(free_apps), np.mean(paid_apps)] # take the mean number of downloads
pyplot.pie(xs, labels=['Free', 'Paid']); # create pie chart
pyplot.title('Average Number of Downloads');
pyplot.legend([int(xs[0]),int(xs[1])], loc='upper left'); # display averages in legend
```

### Result

Based on the previous graph it appears that on average Free apps get far more downloads than Paid apps on the Google Play Store

## Task #3

- compare the Content Rating to the Average number of downloads
- graph relusts

```python
# condese Data Set to just content rating and installs
rated = pd.DataFrame(apps.loc[:,['Installs', 'Content Rating']])
rated.head()
```

```python
# sort by the rating
rated = rated.sort_values(by='Content Rating')
```

```python
# create bar graph of the installs
xs = rated['Content Rating']
ys = [convert(rated.iloc[x,0])for x in range(len(rated))]
pyplot.bar(xs, ys); # plots number of downloads per rating in billions
pyplot.title('Downloads per Rating');
```

### Result

Based on these results, apps rated E, E10+, and T are the most popular, M rated apps are moderatly popular, and Adults only and Unrated apps are not popular at all.
